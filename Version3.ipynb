```python
############### Download elec data from github
url = 'https://raw.githubusercontent.com/Eng-Shady-Hub/Generative_AI_Project_Round3/refs/heads/main/electronics_products_full_edit.xlsx'
elec = pd.read_excel(url, engine="openpyxl")
print(elec.head())
```

```python
elec.info()
elec.shape
```

```python
##### Define Blank 'data_all' dataframe
data_all = pd.DataFrame()
columns = ['product_name','brand','category','sub_category','description']
data_all = data_all.reindex(columns=columns)
```

```python
##################### Combining all data
data_all['product_name']=pd.concat([bigbasket['product'],adidas['Product Name'],amazon['product_name'],flipkart['title'],adidas2['name'],elec['Product_name']], ignore_index=True)
# len(data_all['product_name']) # to confirm lenghth of each column
data_all['brand']=pd.concat([bigbasket['brand'],adidas['Brand'],amazon['brand'],flipkart['brand'],adidas2['brand'],elec['brand']], ignore_index=True)
# len(data_all['brand'])
data_all['category']=pd.concat([bigbasket['category'],adidas['category'],amazon['category'],flipkart['category'],adidas2['category'],elec['category']], ignore_index=True)
# len(data_all['category'])
data_all['sub_category']=pd.concat([bigbasket['sub_category'],adidas['category'],amazon['category'],flipkart['sub_category'],adidas2['breadcrumbs'],elec['sub_category']], ignore_index=True)
# len(data_all['sub_category'])
data_all['description']=pd.concat([bigbasket['description'],adidas['Description'],amazon['about_product'],flipkart['product_details'],adidas2['description'],elec['description']], ignore_index=True)
# len(data_all['description'])
```

```python

data_all = data_all.dropna(subset=["description", "product_name"])
data_all.info()
data_all.shape

# data_all.isnull.sum
```
